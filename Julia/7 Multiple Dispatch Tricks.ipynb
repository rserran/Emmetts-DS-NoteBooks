```julia
elements_tops(v::Vector{Int64}, v2::Vector{Int64}) = [x => y for (x, y) in zip(v, v2)]
```

```julia
mutable struct Category{T <: Any}
    name::String
    Category(name::String, cattype::String = "species") = new{Symbol(cattype)}(name)
    Category{T}(name::String) where {T <: Any} = new{T}(name)
end
```

```julia
human_g = Category("Homo", "genus")
human_sf = Category("Homininae", "family")
human_s = Category("Sapien", "species")
```

```julia
println(typeof(human_g), typeof(human_sf), typeof(human_s))
```

```julia
show(io::IO, cat::Category{:species}) = display(MIME"text/plain"(), "species: $(cat.name)")
show(io::IO, cat::Category{:genus}) = display(MIME"text/plain"(), "genus: $(cat.name)")
```

## Closure Dispatch

```julia
function shower(f::Function)
    ms = methods(f)
    params = [m.sig.parameters[2] for m in ms]
    f(params[1](""))
end
```

```julia
shower() do c::Category{:species}
    println(c)
end
```

## Best Method

```julia
mutable struct OliveDisplay <: AbstractDisplay
    io::IOBuffer
    OliveDisplay() = new(IOBuffer())::OliveDisplay
end
```

```julia
function display(d::OliveDisplay, m::MIME{:olive}, o::Any)
    T::Type = typeof(o)
    mymimes = [MIME"text/html", MIME"text/svg", MIME"image/png",
     MIME"text/plain"]
    mmimes = [m.sig.parameters[3] for m in methods(show, [IO, Any, T])]
    correctm = nothing
    for m in mymimes
        if m in mmimes
            correctm = m
            break
        end
    end
    display(d, correctm(), o)
end

function display(d::OliveDisplay, m::MIME"text/html", o::Any)
    show(d.io, m, o)
end

function display(d::OliveDisplay, m::MIME"image/png", o::Any)
    show(d.io, m, o)
end

function display(d::OliveDisplay, m::MIME"text/plain", o::Any)
    show(d.io, m, o)
end
```

## Awesome Constructors

```julia
abstract type VultaVector end

mutable struct Vec2 <: VultaVector
    x::Int64
    y::Int64
end

mutable struct Vec3 <: VultaVector
    x::Int64
    y::Int64
    z::Int64
end
```

```julia
mutable struct VectorHolder{T <: VultaVector}
    v::Dict{String, T}
    VectorHolder(width::Int64, height::Int64) = new{Vec2}(Dict{String, Vec2}())
    VectorHolder(width::Int64, height::Int64, depth::Int64) = new{Vec3}(Dict{String, Vec3}())
end
```
