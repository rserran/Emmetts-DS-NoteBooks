```julia
wind = svg("sampwind", width = 100, height = 100)
CHANGEME = 2
CHANGEME2 = 10
CHANGEME3 = 7
```

```julia
function draw_space(x::Int64, y::Int64)
    circ = circle("$x-$y", r = 1, cx = x, cy = y)
    if x % CHANGEME == 0 && ~(x % CHANGEME2 == 0)
        style!(circ, "fill" => "red")
    elseif y % CHANGEME2 == 0
        style!(circ, "fill" => "green")
    elseif x % CHANGEME3 == 0 || y % CHANGEME3 == 0
        style!(circ, "fill" => "orange")
    else
        style!(circ, "fill" => "blue")
    end
    push!(wind, circ)
end
```

```julia
current = [0, 0]
while current[2] < 100
    if current[1] == 100
        current[1] = 0
        current[2] += 1
    end
    current[1] += 1
    draw_space(current ...)
end
```

```julia
w, h = 250, 250
window = svg("julia_over", width = w, height = h)
each_w = (w - 10) / 3
each_height = h / 2
circs = [begin
          circ = circle(cx = (each_w * e), cy = each_height, r = 10)
          style!(circ, "fill" => "#$color")
          circ
     end for (e, color) in enumerate(("D5635C", "60AD51", "AA79C1"))]

window[:children] = Vector{AbstractComponent}(circs)
window
```
