```julia
using DataFrames
using Lathe.preprocess: TrainTestSplit
using Lathe.preprocess: StandardScaler, OrdinalEncoder, OneHotEncoder
using Lathe.models: LinearRegression, RandomForestClassifier, Router
```

```julia
x = randn(5000000)
```

```julia
y = randn(5000000)
```

```julia
?(OrdinalEncoder)
```

```julia
function randchoice(opts::Array{String}, count = 5)
    mapping = OrdinalEncoder(opts).lookup
    inverse_dct = Dict([pair[2] => pair[1] for pair in Set(mapping)])
    [inverse_dct[Int64(rand(1:length(opts)))] for i in 1:count]
end
```

```julia
opts = ["A", "B", "C"]
```

```julia
caty = randchoice(opts, 3000000)
```

```julia
catx = randchoice(opts, 3000000)
```

```julia
df = DataFrame(:X => x, :Y => y)
```

```julia
catdf = DataFrame(:CATX => catx, :caty => caty)
```

```julia
@time train, test = TrainTestSplit(df)
```

```julia
tr = :X
te = :Y
trainX = Array(train[!, tr])
trainy = Array(train[!, te])
testX = Array(test[!, tr])
testy = Array(test[!, te])
```

```julia
@time scaler = StandardScaler(trainX)
```

```julia
@time scaled_tX = scaler.predict(trainX)
```

```julia
@time model = LinearRegression(scaled_tX, trainy)
```

```julia
pipe = scaler + model
```

```julia
@time yhat = pipe.predict(testX)
```

```julia
@time train, test = TrainTestSplit(catdf)
```

```julia
tr = :CATX
te = :caty
ctrainX = Array(train[!, tr])
ctrainy = Array(train[!, te])
ctestX = Array(test[!, tr])
ctesty = Array(test[!, te])
```

```julia
@time encoder = OrdinalEncoder(ctrainX)
```

```julia
@time tx_enc = encoder.predict(ctrainX)
```
